```python
#Because we need to make several similar graphs, we are going to make the graphs into functions 
#based on surface reflectance or VI

def VI_graph (x, y, graph_title, point_color):
    #This is the code for our graph, the figure code specifies the graph dimensions
    z = plt.figure(figsize=(100,20))
#This is where you input your x and y variables, and specify the point character, and point color
    plt.plot(x, y, 'o', color= point_color)
#Add the plot title
    plt.title(graph_title)
#This sets the minimum and maximum Y axis values
    plt.ylim(-.2, .2)
#This sets the minimum and maximum X axis values
    plt.xlim('2011-11-20', '2018-02-25')


#This below controls the text sizes

    SMALL_SIZE = 60
    MEDIUM_SIZE = 80
    BIGGER_SIZE = 120


    plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
    plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
    plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
    plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title;
```
