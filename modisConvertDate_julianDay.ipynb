```python
import pandas as pd
```

```python
csv = pd.read_csv('/Volumes/SD/MOD09GAtestRadMean.csv', header = None)
```

```python
csv.head()
```

```python
#Now we will add in column names"
csv.columns = ["file_name", 
               "x_position", 
               "y_position", 
               "state_1km_1",
               "SensorZenith_1",
               "SensorAzimuth_1",
               "SolarZenith_1",
               "SolarAzimuth_1",
               "sur_refl_b01_1",
               "sur_refl_b02_1",
               "sur_refl_b03_1",
               "sur_refl_b04_1",
               "sur_refl_b05_1",
               "sur_refl_b06_1",
               "sur_refl_b07_1",
               "NDVI",
               "EVI",
               "EVI2",
               "QA_result"]
```

```python
#Now check to make sure the column names were inserted
csv.head()
```

```python
#To get the date, we need to split the file_name string

csv[['sensor','year_day','location','tile','julian_day']] = csv['file_name'].str.split('.',expand=True)
```

```python
csv.head()
```

```python
#Now we need to remove the 'A' from the year day column
csv['year_day'] = csv['year_day'].map(lambda x: x.lstrip('A'))
```

```python
csv.head()
```

```python
from datetime import datetime

#First we need to convert the 'date' column to numeric:

csv['year_day'] = csv['year_day'].apply(pd.to_numeric)


#This converts the year_day column from julian day to the common day format

csv['date'] = (pd.to_datetime((csv.year_day// 1000).astype(str)) + pd.to_timedelta(csv.year_day % 1000, unit='D'))
```

```python
#Now we check to make sure that the date looks correct!
csv.head()
```

```python
#We can look at the Datatypes of the files here:
csv.dtypes
```

```python
# To make python understand that "date" is a date, we have to change the data type to datetime
#convert the 'New Date' column to the datetime data type
csv['date'] = pd.to_datetime(csv['date'])
```

```python
#Now select only the columns we want in our output file to use in our graphs

data_for_graphs = csv.loc(axis=1)['file_name',
                    'date',
                    'sur_refl_b01_1',
                    'sur_refl_b02_1',
                    'sur_refl_b03_1',
                    'SensorZenith_1',
                    'SensorAzimuth_1',
                    'SolarZenith_1',
                    'SolarAzimuth_1',
                    'NDVI',
                    'EVI',
                    'EVI2']
```

```python
#Let's take a look at our table to make sure everything is there

data_for_graphs.head()
```

```python
#Now we can export to CSV to save a copy!
data_for_graphs.to_csv("/Volumes/SD/MOD09GA_forGraphs.csv")
```
